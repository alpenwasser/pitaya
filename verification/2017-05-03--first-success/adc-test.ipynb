```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt

%matplotlib inline

df         = pd.read_csv('adc-read-sine--2017-05-03.txt')
df_clipped = pd.read_csv('adc-read-sine--clipped--2017-05-03.txt')


df.VALUE=list(map(lambda x: int(x,16),df.VALUE))
df_clipped.VALUE=list(map(lambda x: int(x,16),df_clipped.VALUE))

plt.figure()
plt.rcParams['figure.figsize']=16,10
df.plot()

plt.figure()
plt.rcParams['figure.figsize']=16,10
df_clipped.plot()
```
